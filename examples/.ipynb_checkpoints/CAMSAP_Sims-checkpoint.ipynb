# Useful Functions

```julia

"""
    X_times_P(X::AbstractMatrix)

# Depends:
- Packages: N/A
- Functions: N/A

A Matrix X times the Left Shift Operator P
"""
function X_times_P(X::AbstractMatrix)
    return circshift(X', -1)'
end

"""
    P_times_X(X::AbstractMatrix)

# Depends:
- Packages: N/A
- Functions: N/A

The Left Shift Operator P times a Matrix X  
"""
function P_times_X(X::AbstractMatrix)
    return circshift(X, 1)
end
```

```julia
"""
    tSVD(X::Matrix, k::Integer)

# Inputs
- X::Matrix
- k::Integer: rank to reduce to

# Outputs
- X_hat
"""
function tSVD(X::Matrix, k::Integer)
    @assert(k > 0, "k is not at least 1")
    @assert(k < minimum(size(X)), "k is too large")
    
    U, s, V = svd(X)
    return (U[:, 1:k] * Diagonal(s[1:k])) * V[:, 1:k]'
end
```

```julia
"""
    top_k_eig(A::Matrix, k::Integer)

# Inputs
- A::Matrix: square
- k::Integer: number of eigenpairs

# Outputs
- w::Vector: eigenvalues, descending magnitude
- Q::Matrix: eigenvectors, columns match q
"""
function top_k_eig(A::Matrix, k::Integer)
    @assert(size(A, 1) == size(A, 2), "A is not square")
    @assert(0 < k, "k is not at least 1")
    @assert(k <= size(A, 1), "k is too large")
    
    w, Q = eigen(A)
    sort_idx = sortperm(abs.(w); rev = true)
    w = w[sort_idx[1:k]]
    Q = Q[:, sort_idx[1:k]]
    
    return w, Q
end
```

```julia
"""
    A_mat(X::Matrix, tau::Integer = 1)

# Inputs
- X::Matrix
- tau::Integer

# Outputs
- X[:, 2:end] * pinv(X[:, 1:(end - 1)])
"""
function A_mat(X::Matrix, tau::Integer = 1)
    @assert(tau >= 1, "tau is too small")
    @assert(tau <= size(X, 2) - 1, "tau is too large")
    
    return X[:, (1 + tau):end] * pinv(X[:, 1:(end - tau)])
end
```

```julia
"""
    permute_IP(Q::Matrix, Q_hat::Matrix)

# Inputs
- Q::Matrix: true vectors
- Q_hat::Matrix: estimate of Q

# Outputs
- Vector of 'best' inner products
"""
function permute_IP(Q::AbstractMatrix, Q_hat::AbstractMatrix)
    k = size(Q, 2)
    
    @assert(k == size(Q_hat, 2), "Q and Q_hat have differing numbers of vectors")
    @assert(size(Q, 1) == size(Q_hat, 1), "Q and Q_hat are not compatible")
    
    # Normalize for unit norm
    Q = mapslices(normalize, Q; dims = 1)
    Q_hat = mapslices(normalize, Q_hat; dims = 1)
    
    # Generate permutations
    pt = permutations(1:k) |> collect
    
    # Find maximum sum of inner products 
    pt_idx = argmax([sum(diag(abs.(Q' * Q_hat[:, x]))) for x in pt])
    pt = pt[pt_idx]
    
    return diag(Q' * Q_hat[:, pt])
end
```

# Cosines with Noise

## Rank 1 Simulations

```julia
# Dimensions
p = 250 # Variables
c_list = 10.0.^(-1.0:0.15:0.4) # c = p / n

# Signal
f = acos(0.5) # Frequency
theta_list = 10.0.^(-1.5:0.15:1.1) # SNR
k = 1 # Rank

# Simulation
trials = 100

# Outputs
cos_rank1_IP = zeros(length(c_list), length(theta_list), trials, 2 * k) * Complex(0, 0)
cos_rank1_IP_S = zeros(length(c_list), length(theta_list), trials, 2 * k) * Complex(0, 0)
cos_rank1_eval = zeros(length(c_list), length(theta_list), trials, 2 * k) * Complex(0, 0)

t1 = time()
for cc in 1:length(c_list)
    c = c_list[cc]
    n = Int(round(p / c)) # Samples
    
    S = normalize(cos.(f * (1.0:n)))
    
    for tt in 1:length(theta_list)
        theta = theta_list[tt]
        
        for tr in 1:trials
            Q = normalize(randn(p))
            
            # Data, Rank-k Approximation
            X = (theta * Q) * S' + randn(p, n) / sqrt(n)
            Xk = tSVD(X, k)
            
            # DMD Matrices
            A = A_mat(X)
            Ak = A_mat(Xk)
            
            # DMD
            what, Qhat = top_k_eig(A, k)
            Shat = mapslices(normalize, (pinv(Qhat) * X)'; dims = 1)
            wkhat, Qkhat = top_k_eig(Ak, k)
            Skhat = mapslices(normalize, (pinv(Qkhat) * Xk)'; dims = 1)
            
            cos_rank1_IP[cc, tt, tr, 1:k] = (Q' * Qhat)[:]
            cos_rank1_IP[cc, tt, tr, (k + 1):(2 * k)] = (Q' * Qkhat)[:]
            cos_rank1_IP_S[cc, tt, tr, 1:k] = (S' * Shat)[:]
            cos_rank1_IP_S[cc, tt, tr, (k + 1):(2 * k)] = (S' * Skhat)[:]
            cos_rank1_eval[cc, tt, tr, 1:k] = what[:]
            cos_rank1_eval[cc, tt, tr, (k + 1):(2 * k)] = wkhat[:]
        end
    end
end
t2 = time()
println(t2 - t1)
```

## Rank 4 Sim

```julia
# Dimensions
p = 250 # Variables
c_list = 10.0.^(-1.0:0.15:0.4) # c = p / n

# Signal
f_list = acos.([0.2; 0.4; 0.6; 0.8]) # Frequency
theta_list = 10.0.^(-1.5:0.15:1.1) # SNR
k = length(f_list) # Rank

# Simulation
trials = 100

# Outputs
cos_rank4_IP = zeros(length(c_list), length(theta_list), trials, 2 * k) * Complex(0, 0)
cos_rank4_IP_S = zeros(length(c_list), length(theta_list), trials, 2 * k) * Complex(0, 0)
cos_rank4_eval = zeros(length(c_list), length(theta_list), trials, 2 * k) * Complex(0, 0)
cos_rank4_sval = zeros(length(c_list), length(theta_list), trials, k)

t1 = time()
for cc in 1:length(c_list)
    c = c_list[cc]
    n = Int(round(p / c)) # Samples
    
    S = mapslices(normalize, cos.((1.0:n) * f_list'); dims = 1)
    
    for tt in 1:length(theta_list)
        theta = theta_list[tt]
        
        for tr in 1:trials
            Q = mapslices(normalize, randn(p, k); dims = 1)
            
            # Data, Rank-k Approximation
            X_clean = (theta * Q) * S' 
            X = X_clean + randn(p, n) / sqrt(n)
            Xk = tSVD(X, k)
            
            # DMD Matrices
            A = A_mat(X)
            Ak = A_mat(Xk)
            
            # DMD
            what, Qhat = top_k_eig(A, k)
            Shat = mapslices(normalize, (pinv(Qhat) * X)'; dims = 1)
            wkhat, Qkhat = top_k_eig(Ak, k)
            Skhat = mapslices(normalize, (pinv(Qkhat) * X)'; dims = 1)
            
            cos_rank4_IP[cc, tt, tr, 1:k] = permute_IP(Q, Qhat)
            cos_rank4_IP[cc, tt, tr, (k + 1):(2 * k)] = permute_IP(Q, Qkhat)
            cos_rank4_IP_S[cc, tt, tr, 1:k] = permute_IP(S, Shat)
            cos_rank4_IP_S[cc, tt, tr, (k + 1):(2 * k)] = permute_IP(S, Skhat)
            cos_rank4_eval[cc, tt, tr, 1:k] = what[:]
            cos_rank4_eval[cc, tt, tr, (k + 1):(2 * k)] = wkhat[:]
            cos_rank4_sval[cc, tt, tr, 1:k] = svdvals(X_clean)[1:k]
        end
    end
end
t2 = time()
println(t2 - t1)
```

# Cosines, sweeping over $\rho$

```julia
# Dimensions
p = 250 # Variables
c_idx = 7
c_list = 10.0.^(-1.0:0.15:0.4) # c = p / n
c = c_list[c_idx]
n = Int(round(p / c)) # Samples

# Signal
f_list = acos.(0.0:0.05:1.0)
theta_list = 10.0.^(-1.5:0.15:1.1) # SNR
k = 1 # Rank

# Simulation
trials = 100

# Outputs
cos_rank1_IP_rho = zeros(length(f_list), length(theta_list), trials, 2 * k) * Complex(0, 0)
cos_rank1_IP_S_rho = zeros(length(f_list), length(theta_list), trials, 2 * k) * Complex(0, 0)
cos_rank1_eval_rho = zeros(length(f_list), length(theta_list), trials, 2 * k) * Complex(0, 0)

t1 = time()
for ff in 1:length(f_list)
    f = f_list[ff]
    S = normalize(cos.(f * (1.0:n)))
    
    for tt in 1:length(theta_list)
        theta = theta_list[tt]
        
        for tr in 1:trials
            Q = normalize(randn(p))
            
            # Data, Rank-k Approximation
            X = (theta * Q) * S' + randn(p, n) / sqrt(n)
            Xk = tSVD(X, k)
            
            # DMD Matrices
            A = A_mat(X)
            Ak = A_mat(Xk)
            
            # DMD
            what, Qhat = top_k_eig(A, k)
            Shat = mapslices(normalize, (pinv(Qhat) * X)'; dims = 1)
            wkhat, Qkhat = top_k_eig(Ak, k)
            Skhat = mapslices(normalize, (pinv(Qkhat) * Xk)'; dims = 1)
            
            cos_rank1_IP_rho[ff, tt, tr, 1:k] = (Q' * Qhat)[:]
            cos_rank1_IP_rho[ff, tt, tr, (k + 1):(2 * k)] = (Q' * Qkhat)[:]
            cos_rank1_IP_S_rho[ff, tt, tr, 1:k] = (S' * Shat)[:]
            cos_rank1_IP_S_rho[ff, tt, tr, (k + 1):(2 * k)] = (S' * Skhat)[:]
            cos_rank1_eval_rho[ff, tt, tr, 1:k] = what[:]
            cos_rank1_eval_rho[ff, tt, tr, (k + 1):(2 * k)] = wkhat[:]
        end
    end
end
t2 = time()
println(t2 - t1)
```

# Rank 2: Cosine and AR(2)

```julia
# Dimensions
p = 250 # Variables
c_list = 10.0.^(-1.0:0.15:0.4) # c = p / n

# Signal

# Cosine
f = acos(0.25) 

# AR(2)
ar = [1.0 / 6.0; 2.0 / 3.0] # Coefficients
ma = []
arma_std = 1.0

theta_list = 10.0.^(-1.5:0.15:1.1) # SNR
k = 2 # Rank
tau = 1 # Lag

trials = 100

# tSVD Outputs
VkPVk = zeros(length(c_list), length(theta_list), trials, k, k)
V_tSVD_IP = zeros(length(c_list), length(theta_list), trials, k)
U_tSVD_IP = zeros(length(c_list), length(theta_list), trials, k)
Q_tSVD_IP = zeros(length(c_list), length(theta_list), trials, k) * Complex(0, 0)
S_tSVD_IP = zeros(length(c_list), length(theta_list), trials, k) * Complex(0, 0)

# tSVD Theory
VkPVk_theory = zeros(length(c_list), length(theta_list), k, k)
V_tSVD_IP2_theory = zeros(length(c_list), length(theta_list), k)
QU_tSVD_IP2_theory = zeros(length(c_list), length(theta_list), k)

# DMD Outputs
Q_DMD_IP = zeros(length(c_list), length(theta_list), trials, k) * Complex(0, 0)
S_DMD_IP = zeros(length(c_list), length(theta_list), trials, k) * Complex(0, 0)

t1 = time()
for cc in 1:length(c_list)
    c = c_list[cc]
    n = Int(floor(p / c)) # Samples
    
    for tt in 1:length(theta_list)
        theta = theta_list[tt]
        
        Theta = [theta; theta / 2.0]
        
        # Theory
        V_tSVD_IP2_theory[cc, tt, :] = max.(0.0, 1.0 .- ((c .+ Theta.^2) ./ (Theta.^2.0 .* (1 .+ Theta.^2.0))))
        QU_tSVD_IP2_theory[cc, tt, :] = max.(0.0, 1.0 .- (c * (1 .+ Theta.^2) ./ (Theta.^2.0 .* (c .+ Theta.^2.0))))
        VkPVk_theory[cc, tt, :, :] = Matrix(Diagonal(V_tSVD_IP2_theory[cc, tt, :] .* [ar[1] / (1.0 - ar[2]); cos(f)]))
        
        Theta = Diagonal(Theta)
        
        for tr in 1:trials
            # Generate Noisy Data
            U = qr(randn(p, k)).Q[:, 1:k]
            V = mapslices(normalize, [gen_arma_sequence(n, ar, ma, arma_std) cos.(f * (1:n))]; dims = 1)
            X = (U * Theta) * V' + (randn(p, n) / sqrt(n))
            
            # tSVD
            Uk, sk, Vk = svd(X)
            Uk = Uk[:, 1:k]
            sk = sk[1:k]
            Vk = Vk[:, 1:k]
            Xk = (Uk * Diagonal(sk)) * Vk'
            
            # Store V_hat^H P V_hat
            VkPVk[cc, tt, tr, :, :] = X_times_P(Vk') * Vk
            
            # Store Inner Products for tSVD
            V_tSVD_IP[cc, tt, tr, :] = permute_IP(V, Vk)
            U_tSVD_IP[cc, tt, tr, :] = permute_IP(U, Uk)
            
            # Compute and Store DMD on tSVD
            A_tSVD = A_mat(Xk, tau)
            w_hat, Q_hat = top_k_eig(A_tSVD, k)
            Q_tSVD_IP[cc, tt, tr, :] = permute_IP(U, Q_hat)
            S_hat = (pinv(Q_hat) * Xk)'
            S_tSVD_IP[cc, tt, tr, :] = permute_IP(V, S_hat)
            
            # Compute and store DMD on Noisy Data
            A_DMD = A_mat(X, tau)
            w_hat, Q_hat = top_k_eig(A_DMD, k)
            Q_DMD_IP[cc, tt, tr, :] = permute_IP(U, Q_hat)
            S_hat = (pinv(Q_hat) * X)'
            S_DMD_IP[cc, tt, tr, :] = permute_IP(V, S_hat)
        end
    end
end
t2 = time()
println(t2 - t1)
```

# Theoretical Conjectures

```julia
function akemann_radius(p, n, r_values)
    if p > n
        pdf = ((p - n) / p) * float.(r_values .== 1.0) + (n / p) * float.(r_values .== 0.0)
    elseif p == n
        pdf = float.(r_values .== 1.0) 
    else
        pdf = (r_values ./ (1.0 .- r_values.^2.0).^2.0) .* (r_values .>= 0.0) .* (r_values .<= sqrt(p / n))
        pdf = 2.0 * ((n / p) - 1) * pdf
    end
    return pdf
end

function akemann_angle(a_values)
    return (1.0 / (2.0 * pi)) * (a_values .<= pi) .* (a_values .>= -pi) 
end
```

```julia
p = 500
n_list = [251; 501; 1001]

trials = 5000
w_list = []
t1 = time()
for nn in 1:length(n_list)
    n = n_list[nn]
    
    w_inner = []
    for tr in 1:trials
        X = randn(p, n) / sqrt(n)
        
        A = X[:, 2:end] * pinv(X[:, 1:(n - 1)])
        push!(w_inner, eigen(A).values)
    end
    push!(w_list, w_inner)
end
t2 = time()
println(t2 - t1)
```

```julia
angle_plots = []
radius_plots = []

nbins = [10 10 30; 20 20 20]

for nn in 1:length(n_list)
    n = n_list[nn]
    r_vals = 0.0:0.01:(1.1 * maximum(abs.(w_list[nn][1])))
    a_vals = (1.1 * minimum(angle.(w_list[nn][1]))):0.01:(1.1 * maximum(angle.(w_list[nn][1])))
    
    p1 = plot(reuse = false, size = (400, 300), title = "Radius, " * L"\frac{p}{n - 1} = " * "$(p / (n - 1))")
    p1 = histogram!(abs.(w_list[nn][1]), bins = nbins[1, nn], label = "Empirical", normalize = true)
    p1 = plot!(r_vals, akemann_radius(p, n - 1, r_vals), color = "red", linewidth = 2, label = "Theory")
    push!(radius_plots, p1)
    
    p1 = plot(reuse = false, size = (400, 300), title = "Phase, " * L"\frac{p}{n - 1} = " * "$(p / (n - 1))")
    p1 = histogram!(angle.(w_list[nn][1]), bins = nbins[2, nn], label = "Empirical", normalize = true)
    p1 = plot!(a_vals, akemann_angle(a_vals), color = "red", linewidth = 2, label = "Theory")
    push!(angle_plots, p1)
end
```

# Hilbert DMD

```julia
# Dimensions
p = 250 # Variables
c_list = 10.0.^(-1.0:0.15:0.4) # c = p / n

# Signal

# Cosine
f = [0.25; 2.0] 

theta_list = 10.0.^(-1.5:0.15:1.1) # SNR
k = 2 # Rank
tau = 1 # Lag

trials = 50

Q_err = zeros(length(c_list), length(theta_list), trials, 3)
S_err = zeros(length(c_list), length(theta_list), trials, 3)

t1 = time()
for cc in 1:length(c_list)
    c = c_list[cc]
    n = Int(floor(p / c)) # Samples
    
    S = mapslices(normalize, cos.((1:n) * f'); dims = 1)
    
    for tt in 1:length(theta_list)
        theta = theta_list[tt]
        
        Theta = [theta; theta / 2.0]
        Theta = Diagonal(Theta)
        
        for tr in 1:trials
            # Generate Noisy Data
            Q = mapslices(normalize, randn(p, k); dims = 1)
            X = (Q * Theta) * S' + (randn(p, n) / sqrt(n))
            
            # tSVD
            Xk = tSVD(X, k)
            # Hilbert
            Xh = mapslices(hilbert, X; dims = 2)
            
            # Form A
            Ak = A_mat(Xk, tau)
            Ah = A_mat(Xh, tau)
            A = A_mat(X, tau)
            
            # Eig
            wk, Qk = top_k_eig(Ak, k)
            wh, Qh = top_k_eig(Ah, k)
            wn, Qn = top_k_eig(A, k)
            
            # S
            Sk = (pinv(Qk) * Xk)'
            Sh = real.((pinv(Qh) * Xh)')
            Sn = (pinv(Qn) * X)'
            
            Q_err[cc, tt, tr, 3] = 2.0 * sum(1.0 .- abs.(permute_IP(Q, Qk)))
            Q_err[cc, tt, tr, 2] = 2.0 * sum(1.0 .- abs.(permute_IP(Q, Qh)))
            Q_err[cc, tt, tr, 1] = 2.0 * sum(1.0 .- abs.(permute_IP(Q, Qn)))
              
            S_err[cc, tt, tr, 3] = 2.0 * sum(1.0 .- abs.(permute_IP(S, Sk)))
            S_err[cc, tt, tr, 2] = 2.0 * sum(1.0 .- abs.(permute_IP(S, Sh)))
            S_err[cc, tt, tr, 1] = 2.0 * sum(1.0 .- abs.(permute_IP(S, Sn)))
        end
    end
end
t2 = time()
println(t2 - t1)
```

```julia
tmp = dropdims(mean(Q_err; dims = 3); dims = 3)

p1 = plot(reuse = false, size = (400, 300), 
    title = "DMD, Q", ylabel = "p / n", xlabel = L"\theta")
p1 = heatmap!(theta_list, c_list, tmp[:, :, 1])


p2 = plot(reuse = false, size = (400, 300), 
    title = "Hilbert-DMD, Q", ylabel = "p / n", xlabel = L"\theta")
p2 = heatmap!(theta_list, c_list, tmp[:, :, 2], yscale = :log10, xscale = :log10)

p3 = plot(reuse = false, size = (400, 300), 
    title = "tSVD-DMD, Q", ylabel = "p / n", xlabel = L"\theta")
p3 = heatmap!(theta_list, c_list, tmp[:, :, 3], yscale = :log10, xscale = :log10)


tmp = dropdims(mean(S_err; dims = 3); dims = 3)

p4 = plot(reuse = false, size = (400, 300), 
    title = "DMD, S", ylabel = "p / n", xlabel = L"\theta")
p4 = heatmap!(theta_list, c_list, tmp[:, :, 1], yscale = :log10, xscale = :log10)

p5 = plot(reuse = false, size = (400, 300), 
    title = "Hilbert-DMD, S", ylabel = "p / n", xlabel = L"\theta")
p5 = heatmap!(theta_list, c_list, tmp[:, :, 2], yscale = :log10, xscale = :log10)

p6 = plot(reuse = false, size = (400, 300), 
    title = "tSVD-DMD, S", ylabel = "p / n", xlabel = L"\theta")
p6 = heatmap!(theta_list, c_list, tmp[:, :, 3], yscale = :log10, xscale = :log10)


plot(p1, p2, p3, p4, p5, p6, layout = (2, 3), size = (1200, 600))
```

# Rank 2: Cosine and AR(2): Missing Data

```julia
# Dimensions
p = 250 # Variables
c_list = 10.0.^[-1.0; 0.4] # 10.0.^(-1.0:0.15:0.4) # c = p / n

# Signal

# Cosine
f = acos(0.25) 

# AR(2)
ar = [1.0 / 6.0; 2.0 / 3.0] # Coefficients
ma = []
arma_std = 1.0

theta_list = 10.0.^(-1.5:0.15:1.1) # SNR
q_list = 10.0.^(-1.5:0.15:0.0) # P(obs)
k = 2 # Rank
tau = 1 # Lag

trials = 50

# tSVD Outputs
VkPVk = zeros(length(c_list), length(theta_list), length(q_list), trials, k, k)
V_tSVD_IP = zeros(length(c_list), length(theta_list), length(q_list), trials, k)
U_tSVD_IP = zeros(length(c_list), length(theta_list), length(q_list), trials, k)
Q_tSVD_IP = zeros(length(c_list), length(theta_list), length(q_list), trials, k) * Complex(0, 0)
S_tSVD_IP = zeros(length(c_list), length(theta_list), length(q_list), trials, k) * Complex(0, 0)

# tSVD Theory
VkPVk_theory = zeros(length(c_list), length(theta_list), length(q_list), k, k)
V_tSVD_IP2_theory = zeros(length(c_list), length(theta_list), length(q_list), k)
QU_tSVD_IP2_theory = zeros(length(c_list), length(theta_list), length(q_list), k)

# DMD Outputs
Q_DMD_IP = zeros(length(c_list), length(theta_list), length(q_list), trials, k) * Complex(0, 0)
S_DMD_IP = zeros(length(c_list), length(theta_list), length(q_list), trials, k) * Complex(0, 0)

t1 = time()
for cc in 1:length(c_list)
    c = c_list[cc]
    n = Int(floor(p / c)) # Samples
    
    for tt in 1:length(theta_list)
        theta = theta_list[tt]
        
        Theta = [theta; theta / 2.0]
        Theta = Diagonal(Theta)
        
        for qq in 1:length(q_list)
            q = q_list[qq]
            Theta_q = [theta; theta / 2.0] * sqrt(q) # Note: assigning Theta * sqrt(q) causes weird issues

            # Theory
            V_tSVD_IP2_theory[cc, tt, qq, :] = max.(0.0, 1.0 .- ((c .+ Theta_q.^2) ./ (Theta_q.^2.0 .* (1 .+ Theta_q.^2.0))))
            QU_tSVD_IP2_theory[cc, tt, qq, :] = max.(0.0, 1.0 .- (c * (1 .+ Theta_q.^2) ./ (Theta_q.^2.0 .* (c .+ Theta_q.^2.0))))
            VkPVk_theory[cc, tt, qq, :, :] = Matrix(Diagonal(V_tSVD_IP2_theory[cc, tt, qq, :] .* [ar[1] / (1.0 - ar[2]); cos(f)]))

            for tr in 1:trials
                # Generate Noisy Data
                U = qr(randn(p, k)).Q[:, 1:k]
                V = mapslices(normalize, [gen_arma_sequence(n, ar, ma, arma_std) cos.(f * (1:n))]; dims = 1)
                X = ((U * Theta) * V' + (randn(p, n) / sqrt(n))) .* float.(rand(p, n) .<= q)
                
                # tSVD
                Uk, sk, Vk = svd(X)
                Uk = Uk[:, 1:k]
                sk = sk[1:k]
                Vk = Vk[:, 1:k]
                Xk = (Uk * Diagonal(sk)) * Vk'

                # Store V_hat^H P V_hat
                VkPVk[cc, tt, qq, tr, :, :] = X_times_P(Vk') * Vk

                # Store Inner Products for tSVD
                V_tSVD_IP[cc, tt, qq, tr, :] = permute_IP(V, Vk)
                U_tSVD_IP[cc, tt, qq, tr, :] = permute_IP(U, Uk)

                # Compute and Store DMD on tSVD
                A_tSVD = A_mat(Xk, tau)
                w_hat, Q_hat = top_k_eig(A_tSVD, k)
                Q_tSVD_IP[cc, tt, qq, tr, :] = permute_IP(U, Q_hat)
                S_hat = (pinv(Q_hat) * Xk)'
                S_tSVD_IP[cc, tt, qq, tr, :] = permute_IP(V, S_hat)

                # Compute and store DMD on Noisy Data
                A_DMD = A_mat(X, tau)
                w_hat, Q_hat = top_k_eig(A_DMD, k)
                Q_DMD_IP[cc, tt, qq, tr, :] = permute_IP(U, Q_hat)
                S_hat = (pinv(Q_hat) * X)'
                S_DMD_IP[cc, tt, qq, tr, :] = permute_IP(V, S_hat)
            end
        end
    end
end
t2 = time()
println(t2 - t1)
```

```julia
function gen_missing_IP_plots(data, c_list, theta_list, q_list, var_str, alg_str)
    m_data = dropdims(mean(abs2.(data); dims = 4); dims = 4)
    
    if "v" == var_str
        title_str = L"\left|\widehat{v}^H v\right|^2"
    elseif "u" == var_str
        title_str = L"\left|\widehat{u}^H u\right|^2"
    elseif "q" == var_str
        title_str = L"\left|\widehat{q}^H u\right|^2"
    elseif "s" == var_str
        title_str = L"\left|\widehat{s}^H v\right|^2"
    end
    
    if "dmd" == alg_str
        title_str = title_str * ", DMD"
    elseif "tsvd" == alg_str
        title_str = title_str * ", tSVD"
    end
    title_str = title_str * "\nc = "
    
    c_str = "1e" .* string.(log10.(c_list))
    
    plot_arr = []
    for c_idx in 1:length(c_list)
        for idx in 1:size(m_data, 4)
            title_inner = title_str * c_str[c_idx] * ", Index " * string(idx)
            p1 = plot(reuse = false, size = (400, 300), clims = (0, 1), xlims = (q_list[1], q_list[end]),
                title = title_inner, xlabel = L"q = P(obs)", ylabel = L"\theta = \theta_1 = 2 \theta_2"
            )
            p1 = heatmap!(q_list, theta_list, m_data[c_idx, :, :, idx], yscale = :log10, xscale = :log10)
            if "tsvd" == alg_str
                if 1 == idx
                    c_scale = 1.0
                else
                    c_scale = 2.0
                end
                q_pl = [q_list[1] / 2.0; q_list; q_list[end] * 1.1]
                p1 = plot!(q_pl, c_scale * c_list[c_idx]^0.25 ./ q_pl.^0.5, label = "", color = "white", linewidth = 2)
            end
            push!(plot_arr, p1)
        end
    end
    
    return plot_arr
end

plots_u = gen_missing_IP_plots(U_tSVD_IP, c_list, theta_list, q_list, "u", "tsvd")
plots_v = gen_missing_IP_plots(U_tSVD_IP, c_list, theta_list, q_list, "v", "tsvd")
plots_udmd = gen_missing_IP_plots(Q_tSVD_IP, c_list, theta_list, q_list, "q", "tsvd")
plots_vdmd = gen_missing_IP_plots(S_tSVD_IP, c_list, theta_list, q_list, "s", "tsvd")
plots_q = gen_missing_IP_plots(Q_DMD_IP, c_list, theta_list, q_list, "q", "dmd")
plots_s = gen_missing_IP_plots(S_DMD_IP, c_list, theta_list, q_list, "s", "dmd")
```

```julia
VPV_err = sum([dropdims(sum(abs2.(VkPVk[:, :, :, idx, :, :] - VkPVk_theory); dims = (4, 5)); dims = (4, 5)) for idx in 1:trials]) / trials
VPV_offdiag = sum([VkPVk[:, :, :, idx, 2, 1].^2 + VkPVk[:, :, :, idx, 1, 2].^2 for idx in 1:trials] / trials)
plots_vpv = []

c_str = "1e" .* string.(log10.(c_list))

title_str = L"||\widehat{V}^H P \widehat{V} - diag\left(\widetilde{\alpha}_i^2 \gamma_i\right)||" * "\nc = "
for c_idx in 1:length(c_list)
    title_inner = title_str * c_str[c_idx] 
    p1 = plot(reuse = false, size = (400, 300), xlims = (q_list[1], q_list[end]),
        title = title_inner, xlabel = L"q = P(obs)", ylabel = L"\theta = \theta_1 = 2 \theta_2"
    )
    p1 = heatmap!(q_list, theta_list, VPV_err[c_idx, :, :], yscale = :log10, xscale = :log10)
    push!(plots_vpv, p1)
end

title_str = L"||off\,diag\left(\widehat{V}^H P \widehat{V}\right)||" * "\nc = "
for c_idx in 1:length(c_list)
    title_inner = title_str * c_str[c_idx] 
    p1 = plot(reuse = false, size = (400, 300), xlims = (q_list[1], q_list[end]),
        title = title_inner, xlabel = L"q = P(obs)", ylabel = L"\theta = \theta_1 = 2 \theta_2"
    )
    p1 = heatmap!(q_list, theta_list, VPV_offdiag[c_idx, :, :], yscale = :log10, xscale = :log10)
    push!(plots_vpv, p1)
end
```
